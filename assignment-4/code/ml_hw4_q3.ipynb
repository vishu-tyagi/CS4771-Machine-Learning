```python
import os
from pathlib import Path

os.chdir('/content/drive/MyDrive/ML/hw4')
IMAGES = Path.cwd() / 'images'

from kmeans import *
from utils import *

import numpy as np
import pandas as pd

from sklearn.datasets import make_circles, make_moons
```

```python
# load data
xdev1, ydev1 = make_circles(n_samples=1000, random_state=64)

# run k-means
preds1, means1 = kmeans(X=xdev1, k=2, iters=100, return_means=True)

# plot results
plot(X=xdev1, Y=ydev1, k_means_clusters=preds1, k_means_centers=means1, method='k-means', save_loc=IMAGES/'concentric.png')
```

```python
# load data
x_small, y_small = make_circles(n_samples=1000, random_state=3, noise=0, factor = 0.7)
x_large, y_large = make_circles(n_samples=1000, random_state=3, noise=0, factor = 0.4)
y_large[y_large ==1] = 2

df = pd.DataFrame(np.vstack([x_small,x_large]),columns=['x1','x2'])
df['label'] = np.hstack([y_small, y_large])

xdev2 = np.array(df[['x1', 'x2']])
ydev2 = df.label

# run k-means
preds2, means2 = kmeans(X=xdev2, k=2, iters=100, return_means=True)

# plot results
plot(X=xdev2, Y=ydev2, k_means_clusters=preds2, k_means_centers=means2, method='k-means', save_loc=IMAGES/'three_concentric.png')
```

```python
# load data
xdev3, ydev3 = make_moons(n_samples=1000, random_state=64)

# run k-means
preds3, means3 = kmeans(X=xdev3, k=2, iters=100, return_means=True)

# plot results
plot(X=xdev3, Y=ydev3, k_means_clusters=preds3, k_means_centers=means3, method='k-means', save_loc=IMAGES/'moons.png')
```

```python
n_neighbors = 12

# run spectral k-means
nn_preds1, nn_means1 = spectral_kmeans(X=xdev1, k=2, n_neighbors=n_neighbors, lowest_r=2, iters=100, return_means=True)

# plot results
plot(X=xdev1, Y=ydev1, k_means_clusters=nn_preds1, k_means_centers=nn_means1, method='Spectral k-means', n_neighbors=n_neighbors, save_loc=IMAGES/'spectral_concentric.png')
```
